```python
import pandas as pd
from pathlib import Path
```

```python
PROJECT_ROOT = Path.cwd().parent

RAW_DATA_PATH = PROJECT_ROOT / "dataset" / "train_FD001.txt"
CSV_DATA_PATH = PROJECT_ROOT / "dataset" / "train_FD001.csv"

print("Raw TXT path :", RAW_DATA_PATH)
print("CSV output path :", CSV_DATA_PATH)
```

```python
df = pd.read_csv(
    RAW_DATA_PATH,
    sep=r"\s+",     # raw string → avoids SyntaxWarning
    header=None
)

df.head()
```

```python
columns = (
    ["engine_id", "cycle"] +
    [f"op_setting_{i}" for i in range(1, 4)] +
    [f"sensor_{i}" for i in range(1, 22)]
)

df.columns = columns
df.head()
```

```python
df.to_csv(CSV_DATA_PATH, index=False)
print("✅ CSV file successfully saved at:", CSV_DATA_PATH)
```
